```python
class Database:
    """This class will hold the login details of registered users. It will also house the login prompt(function)."""
    def __init__(self, fullname):
        self.fullname = fullname
        
    def login_saver(self, fullname, username, password, pin):
        """The username will be the user's email. I should later modify this method to only accept login parameters, scan database
        (when, in the future, database has the ability to retain previous info passed to it) to see if any of the parameters 
        already existed. It would return a value to the appropriate method which will use it to determine the next steps. 
        Concerned method (in NewUser Class) is 'Withdrawal_pin_setup'"""
        self.fullname = fullname
        self.username = username
        self.password = password
        self.pin = pin
        login_file = open(r"C:\Users\Gabriel O Nnamoko\Desktop\BetZazzy Database\Login Details.txt", 'r')
        for lines in login_file:
            if self.username in lines:
                print("\nEmail address has been registered with another account. Please use a new one")
                break
        if self.username not in login_file:
            in_file = 0
        else:
            in_file = 1
        login_file.close()
        if in_file == 0:
            with open(r"C:\Users\Gabriel O Nnamoko\Desktop\BetZazzy Database\Login Details.txt", 'a') as login_file:
                login_file.write(self.fullname+ '|' + self.username + '|' + self.password + '\n')
                print("\nLogin details saved successfully.")
            with open(r"C:\Users\Gabriel O Nnamoko\Desktop\BetZazzy Database\Withdrawal Pins.txt", 'a') as pin_file:
                pin_file.write(self.username + '|' + self.user_pin + '\n')
        
    def login(self):
        """The username here will be the user's email address too, to mitigate fraud and establish who actually owns
        an account in an instance where the withdrawal pin has been forgotten."""
        self.line_holder = ''
        self.username = input("\nEnter your email: ")
        with open(r"C:\Users\Gabriel O Nnamoko\Desktop\BetZazzy Database\Login Details.txt", 'r') as f:
            for lines in f:
                if self.username in lines:
                    self.line_holder = lines
                    break
            if self.line_holder != '':
                self.line_holder = self.line_holder.split(sep='|')
                #self.expected_password = self.line_holder[2]
                print(self.line_holder[2])
                self.enter_password = input(f"\nWelcome {self.line_holder[0]}\n\tPlease enter your password: ")
                if self.enter_password == self.line_holder[2]:
                    print(f"\n{self.line_holder[0]} you are successfully logged in.")
                else:
                    while self.enter_password != self.line_holder[2]:
                        self.enter_password = input(f"\n{self.line_holder[2]}\n ")
                    else:
                        print(f"\n{self.line_holder[2]} you are successfully logged in.")
                    
                    
            else:
                print("\nThis email address is not linked to any account.")
        

class NewUser(Database):
    company_name = 'BetZazzy'
    supported_payment_services = ['Paypal', 'Local Bank']
    def __init__(self, fullname, age):
        super().__init__(fullname)
        self.age = age
        if self.age < 18:
            print(f"{self.fullname} you will not be able to continue with {NewUser.company_name} as you are below the legal age.")
        else:
            NewUser.complete_registration(self)
            
    def complete_registration(self):
        import string
        upper_case_letters = 0
        numbers_in_password = 0
        special_characters = 0
        lower_case_characters = 0
        unknown_characters = 0
        self.country = input("\nEnter your country: ")
        self.contact_address = input("\nEnter your full mailing address: \n")
        self.phone_number = input("\nEnter your phone number (with country code): ")
        self.prompt = 'Setup your login details'
        print(f"\n{self.prompt:>16}\n")
        self.user_email_address = input("\nEnter your functional email address: ")
        while '@' and '.com' not in self.user_email_address:
            self.user_email_address = input("\nPlease enter a valid email address: ")
        self.user_password = input("\nCreate a password: ")
        self.password_length = len(self.user_password)
        for chars in self.user_password:
            try:
                if chars.islower():
                    lower_case_characters += 1
                elif chars.isupper():
                    upper_case_letters += 1
                elif chars.isdigit():
                    numbers_in_password += 1
                elif chars in string.punctuation:
                    special_characters += 1
            except TypeError:
                unknown_characters += 1
                continue
        while self.password_length < 8 or lower_case_characters < 1 or upper_case_letters < 1 or numbers_in_password < 1 or special_characters < 1 or unknown_characters >= 1:
            self.user_password = input("\nPassword must not be less than 8 characters (uppercase, lowercase, numbers and special characters only) \n\nCreate a password:")
            self.password_length = len(self.user_password)
            for chars in self.user_password:
                try:
                    if chars.islower():
                        lower_case_characters += 1
                    elif chars.isupper():
                        upper_case_letters += 1
                    elif chars.isdigit():
                        numbers_in_password += 1
                    elif chars in string.punctuation:
                        special_characters += 1
                except TypeError:
                    unknown_characters += 1
                    continue
        NewUser.setup_bank_account_details(self)
        
    def setup_bank_account_details(self):
        counter = 1
        print("\nChoose any of the following services that you want to use:")
        for items in NewUser.supported_payment_services:
            print(str(counter)+'.', items)
            counter += 1
        self.user_input = int(input())
        if self.user_input == 1:
            self.user_paypal_email = input("\nEnter your Paypal email address: ")
            while '@' and '.com' not in self.user_paypal_email:
                self.user_paypal_email = input("\nEnter a valid email address: ")
            else:
                print("\nWithdrawal option successfully set up.")
                NewUser.set_up_withdrawal_pin(self)
        elif self.user_input == 2:
            self.user_bank_name = input("\nEnter your bank's name: ")
            self.user_account_number = input("\nEnter account number: ")
            print("\nWithdrawal option successfully set up.")
            NewUser.set_up_withdrawal_pin(self)
        else:
            print("\nSuspicious activity. Session terminated.")
            
    def set_up_withdrawal_pin(self):
        upper_case_letters = 0
        numbers_in_pin = 0
        special_characters = 0
        lower_case_characters = 0
        unknown_characters = 0
        self.counter = 0
        print("\nSet up a 4 digit withdrawal pin:\n")
        self.user_pin = input('\t')
        self.pin_length = len(self.user_pin)
        for chars in self.user_pin:
            try:
                if chars.islower():
                    lower_case_characters += 1
                elif chars.isupper():
                    upper_case_letters += 1
                elif chars.isdigit():
                    numbers_in_pin += 1
                elif chars in string.punctuation:
                    special_characters += 1
            except TypeError:
                unknown_characters += 1
                continue
        while self.pin_length > 4 and lower_case_characters >= 1 or upper_case_letters >= 1 or special_characters >= 1 or unknown_characters >= 1:
            self.user_pin = input("\nPin must not be greater than 4 characters (digits ONLY) \n\nCreate a withdrawal pin:")
            self.pin_length = len(self.user_pin)
            for chars in self.user_pin:
                try:
                    if chars.islower():
                        lower_case_characters += 1
                    elif chars.isupper():
                        upper_case_letters += 1
                    elif chars.isdigit():
                        numbers_in_pin += 1
                    elif chars in string.punctuation:
                        special_characters += 1
                except TypeError:
                    unknown_characters += 1
                    continue
            if self.pin_length == 4 and numbers_in_pin == 4:
                print("\nWithdrawal pin set successfully")
                break
        
        else:
            print("\nWithdrawal pin set successfully")
        Database.login_saver(self, self.fullname, self.user_email_address, self.user_password, self.user_pin)
```
